# Amazon Stock Close Price Prediction

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
from tensorflow import keras
```

## 2. Import Dataset and edit columns

```python
stocks = pd.read_csv('test_dataset.csv')
stocks = stocks.drop([0, 1])
stocks.rename(columns={'Price': 'Date'}, inplace=True)
stocks = stocks.drop(columns=['Adj Close','Date'])
stocks.head()
```

## 3. Feature Engineering
2 More features, Price Range and Volume Range, 
where "Price Range" column means what was the opening / closing price range for each record

```python
numeric_columns = ['Close', 'High', 'Low', 'Open', 'Volume']
stocks[numeric_columns] = stocks[numeric_columns].astype(float)
stocks['Price_Range'] = stocks['High'] - stocks['Low']
stocks['Volume_Range'] = stocks['Volume'] / (stocks['High'] - stocks['Low'])
print(numeric_columns)
```

## 4. Data Analysis

```python
stocks.info() 
```

```python
stocks.describe()
```

```python
import seaborn as sns

sns.lineplot(data=stocks, x='High', y='Close')
# sns.barplot(data=stocks, x='Close', y='Date')
```

## 4. Data preprocessing

```python
stocks_X = stocks[numeric_columns].drop(columns=['Close'])
stocks_y = stocks['Close']
print("Dataset (records, features) : ", stocks_X.shape)
print("Dataset Labels (records,) : ",stocks_y.shape)
```

### 4.1 Training set and Testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(stocks_X, stocks_y, test_size=0.2, random_state=42)
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

## 4.2 Data Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("X_train After Scaling:")
df = pd.DataFrame(X_train, columns=numeric_columns[1:])
df.describe()
# df2 = pd.DataFrame(X_test, columns=numeric_columns[1:])
# df2.head()
```

# 5. Model Building
using keras "Dense", "Input" layers to build Multi-layer Proceptron (MLP)

```python
from keras.layers import Dense, Input
from keras.models import Sequential

model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='linear'),
])
```

### 5.1 Model Configurations
using Adam optimizer and (MSE) as loss calculation method and (MAE - MAPE) as metrics to calculate model accuracy

```python
from keras.optimizers import Adam, SGD
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mean_absolute_error', 'mape'])
```

### 5.2 Model Training

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, batch_size=8)
```

## 6. Model Evaluating

```python
score = model.evaluate(X_test, y_test, verbose=0)
loss , mae, mape = score
print('Model Performance in Testing Set')
print('='*10)
print('Model Loss: ', loss)
print('Model Mean absolute error: ', mae)
print('Model Mean absolute percentage error: ', mape)
```

## 6.1 Model Performance Visualize

```python
import matplotlib.pyplot as plt
sns.lineplot(data=history.history['loss'])
sns.lineplot(data=history.history['val_loss'])
plt.show()
```

# Underfitting Problem
## Make the model simple

```python
underfit_model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(4, activation='relu'),
    Dense(1, activation='linear')
])
```

## Less epochs, more batch size, more simple model => Underfitting

```python
underfit_model.compile(loss='mean_squared_error', metrics=['mean_absolute_error', 'mape'])
```

```python
history = underfit_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32)
```

```python
underfit_model.evaluate(X_test, y_test)
```

```python
sns.lineplot(data=history.history['loss'])
sns.lineplot(data=history.history['val_loss'])
plt.title('Underfitting Learning Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

# Overfitting Problem
## Make model complex

```python
overfit_model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dense(20, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='linear')
])
```

## More Epochs, less batch size, without regularization => Overfitting

```python
overfit_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mean_absolute_error', 'mape'])
```

```python
history = overfit_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=1000, batch_size=8)
```

```python
overfit_model.evaluate(X_test, y_test)
```

```python
sns.lineplot(data=history.history['loss'])
sns.lineplot(data=history.history['val_loss'])
plt.title('Overfitting Learning Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```
